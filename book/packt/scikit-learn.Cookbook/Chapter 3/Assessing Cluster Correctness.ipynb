#Getting Ready

We've talk a little bit about assessing clusters when the ground truth is not known.  However we've not yet talked about how assessing kmeans when the cluster is known.  In a lot of cases this isn't knowable, however if there is outside annotation sometimes we will known the ground truth - or at least proxy.

So then, we'll live in the world where we have some outside agent supplying us with the ground truth.

We'll create a simple dataset then evaluate the measures of correctness against the ground truth in several ways and discuss.

```python
from sklearn import datasets
from sklearn import cluster
```

```python
blobs, ground_truth = datasets.make_blobs(1000, centers=3, cluster_std=2)
```

```python
f, ax = plt.subplots(figsize=(7, 5))

colors = ['r', 'g', 'b', 'y']

for i in range(4):
    p = blobs[ground_truth == i]

    ax.scatter(p[:,0], p[:,1], c=colors[i], label="Cluster {}".format(i))
    
ax.set_title("Cluster With Ground Truth")
ax.legend()
```

```python
kmeans = cluster.KMeans(n_clusters=3)
```

```python
kmeans.fit(blobs)
```

```python
kmeans.cluster_centers_
```

```python
f, ax = plt.subplots(figsize=(7, 5))

colors = ['r', 'g', 'b', 'y']

for i in range(3):
    p = blobs[ground_truth == i]

    ax.scatter(p[:,0], p[:,1], c=colors[i], label="Cluster {}".format(i))
    
ax.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, color='black', label='Centers')
ax.set_title("Cluster With Ground Truth")
ax.legend()
```

As you can see the red and the yellow cluster are mostly correct, or at least more correct by comparison to the green and blue.

Now that we can view the clustering performance as a classification exercise the same metrics which are useful in that context are useful here.

```python
for i in range(3):
    print (kmeans.labels_ == ground_truth)[ground_truth == i].astype(int).mean()
```

Clearly the we have some backwards clusters.  So let's get that straigtened out then we'll look at the accuracy.

```python
new_ground_truth = ground_truth.copy()
```

```python
new_ground_truth[ground_truth == 0] = 2
new_ground_truth[ground_truth == 2] = 0
```

```python
for i in range(3):
    print (kmeans.labels_ == new_ground_truth)[ground_truth == i].astype(int).mean()
```

So we're roughly correct 90% of the time.

The second measure of similarity we'll look at is mutual information score.

```python
from sklearn import metrics
```

```python
metrics.normalized_mutual_info_score(ground_truth, kmeans.labels_)
```

As the score tends to 0 the label assignments are probably not generated through similar processes, however closer to 1 means that there is a large amount of agreement between the two labels.

For example, let's look at what happens when the mutual information score scores itself.

```python
metrics.normalized_mutual_info_score (ground_truth, ground_truth)
```

Given the name, we can tell that there is probably an unnormalized mutual_info_score.

```python
metrics.mutual_info_score(ground_truth, kmeans.labels_)
```

These are very close, however normalized mututal information is the mutual information dividied by the root of product of the entropy of each set truth and assigned labels.

# There's More

One cluster metric we haven't talk about yet that is _not_ reliant on the ground truth is intertia.  Is not very well documented as a metric at the moment.  But it is the metric that KMeans is minimizing.

Inertia is the sum of squared difference between each point and it's assigned cluster.  We can use a little numpy to determine this.

```python
kmeans.inertia_ 
```

Inutively, it makes sense that lower values for intertia are better.
